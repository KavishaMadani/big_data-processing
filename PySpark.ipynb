```python
import numpy as np
from pyspark.sql import SparkSession
```

```python
# Create a Spark session
spark = SparkSession.builder.appName("RDD_Task_1").getOrCreate()
```

```python
# Generate 100 random numbers in range 0 to 10 using numpy
np.random.seed(7)
random_numbers = np.random.randint(0, 11, 100)
random_numbers
```

```python
# Create RDD from the generated data
rdd_numbers = spark.sparkContext.parallelize(random_numbers)
```

```python
frequency = rdd_numbers.countByValue()
```

```python
for num, freq in sorted(frequency.items()):
    print(f"Number {num}: {freq} occurrences")
```

```python
# Stop the Spark session
spark.stop()
```
