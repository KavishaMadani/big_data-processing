```python
html_content = response.content
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
soup
```

```python
links = soup.find_all("a",attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
links
```

```python
links = soup.find_all("span", attrs = {'class' : 'a-size-medium a-color-base a-text-normal'})
links
```

```python
for i in links:
    print(i.text)
```

```python
names=[]
for i in links:
    names.append(i.text)
names=np.transpose(names)
print(names)
```

```python
links1 = soup.find_all("i", attrs = {'class' : "a-icon a-icon-star-small a-star-small-4-5 aok-align-bottom"})
links1
```

```python
ratings=[]
for i in soup.find_all("i", attrs = {'class' : "a-icon a-icon-star-small a-star-small-4-5 aok-align-bottom"}):
    ratings.append(i.text.strip())
ratings=np.transpose(ratings)
print(ratings)
```

```python
bought=[]
for i in soup.find_all("span", attrs = {'class' : "a-size-base a-color-secondary"}):
    bought.append(i.text.strip())
bought=np.transpose(bought)
print(bought)
```

```python
price=[]
for i in soup.find_all("span", attrs = {'class' : "a-offscreen"}):
    price.append(i.text.strip())
price=np.transpose(price)
print(price)
```

```python
data_dict = {'Names': names, 'Ratings': ratings, 'Bought': bought, 'Price': price}
```

```python
min_length = min(len(names), len(ratings), len(bought), len(price))
names = names[:min_length]
ratings = ratings[:min_length]
bought = bought[:min_length]
price = price[:min_length]
```

```python
data = pd.DataFrame(data_dict)
```

```python
data.replace('', np.nan, inplace=True)
```
